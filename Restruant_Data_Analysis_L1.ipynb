## Task 3: Price Range Distribution

### Objective

To analyze the distribution of price ranges among restaurants and calculate the percentage of restaurants in each price category.


```python
price_data = df['Price range'].value_counts()
price_data = price_data.sort_index()
price_data.plot(kind='bar', width=0.6)
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
price_percent = df['Price range'].value_counts(normalize=True) * 100
price_percent = price_percent.round(2)
print("Percentage of restaurants:\n")
for i in price_percent.index:
    print("Range", i, "-", str(price_percent[i]) + "%")
```

### Conclusion

In this task, I visualized the distribution of price ranges using a bar chart.  
I also calculated the percentage of restaurants in each category, where Range 1 has the highest share (46.51%) and Range 4 the lowest (6.14%).


## Task 4: Online Delivery

### Objective

To determine the percentage of restaurants that offer online delivery and compare the average ratings of restaurants with and without online delivery.


```python
data = df['Has Online delivery'].value_counts(normalize=True) * 100
data = data.round(2)
print("Percentage of restaurants with online delivery: ",str(data['Yes']) + "%")
```

```python
ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()
ratings = ratings.round(2)
print("Average ratings:\n","Offline rating :", ratings['No'],"\n","Online rating  :", ratings['Yes'])
```

### Conclusion

The percentage of restaurants that offer online delivery is 25.69%.  
The average rating of restaurants with online delivery (3.25) is higher than those without online delivery (2.46).
